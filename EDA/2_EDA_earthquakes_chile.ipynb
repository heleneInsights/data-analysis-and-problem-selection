#### Outlier Detection

```python
# Select numeric columns
numeric_columns = df.select_dtypes(
    include=np.number
).columns

if len(numeric_columns) > 0:

    # create boxplots
    # Dashboard layout
    n_cols = 4
    n_rows = math.ceil(len(numeric_columns) / n_cols)

    # Create subplot figure
    fig = make_subplots(
        rows=n_rows,
        cols=n_cols,
        subplot_titles=numeric_columns
    )

    # Add boxplots
    for i, col in enumerate(numeric_columns):
        row = (i // n_cols) + 1
        col_pos = (i % n_cols) + 1

        fig.add_trace(
            go.Box(
                y=df[col],
                name=col,
                boxmean=True
            ),
            row=row,
            col=col_pos
        )

    # Update layout
    fig.update_layout(
        title="Interactive Boxplot Dashboard — Numeric Features",
        height=300 * n_rows,
        width=1400,
        showlegend=False,
        template="plotly_white"
    )

    # Add axis labels
    fig.update_yaxes(title_text="Observed Values")

    # Show dashboard
    fig.show()

else:
    print("No numeric variables available for outlier analysis.")
```

**Analysis**

All numeric variables contain observations identified as outliers according to the IQR rule.

However, the percentage of outliers is relatively low, ranging from 0.20% for `Latitude` to 2.69% for `Magnitude`.

Since these values correspond to real earthquake measurements, they should not automatically be considered errors. Extreme magnitudes, depths, or geographic locations may represent significant seismic events and could provide valuable information for the analysis.

If required for modeling, transformations such as logarithmic scaling or robust scaling could be evaluated to reduce the influence of extreme values while preserving the original observations.

IQR calculation.

```python
feature_columns = df.select_dtypes(
    include=np.number
).columns

outlier_summary = []

for col in feature_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    n_outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()

    outlier_summary.append({
        "Feature": col,
        "Outliers": n_outliers,
        "Percentage": round(n_outliers / len(df) * 100, 2)
    })

outlier_df = pd.DataFrame(outlier_summary)
outlier_df
```

### Heatmap visualization

```python
# Correlation matrix
corr_matrix = df[numeric_columns].corr()

# Create mask for upper triangle
mask = np.triu(
    np.ones_like(corr_matrix, dtype=bool)
)

# Apply mask
corr_matrix_masked = corr_matrix.mask(mask)

fig = px.imshow(
    corr_matrix_masked,
    text_auto=".2f",
    color_continuous_scale="BuGn",
    aspect="auto",
    title="Correlation Matrix Heatmap"
)

fig.update_layout(
    width=900,
    height=800
)

fig.show()
```

**Analysis**

`Latitude` and `Longitude` show a relatively strong positive correlation, reflecting the geographic distribution of earthquakes along Chile. `Depth` also exhibits a moderate positive correlation with `Longitude`, suggesting that earthquake depth may vary across different seismic regions.

`Latitude` has a moderate correlation with `Depth` (0.38), indicating that some regions may experience deeper earthquakes than others.

`Magnitude` shows very weak correlations with all other variables, suggesting that earthquake intensity is not strongly associated with location or depth in a simple linear relationship.

These results indicate that geographic variables may be useful for identifying spatial patterns of seismic activity, while earthquake magnitude may require more advanced analysis to uncover meaningful relationships.

## Geospatial Visualization

```python
# Create copy for geospatial analysis
df_geo = df_datetime.copy()

# Slider
min_magnitude = widgets.FloatSlider(
    value=5.0,
    min=float(df_geo["Magnitude"].min()),
    max=float(df_geo["Magnitude"].max()),
    step=0.1,
    description="Min Mag:",
    continuous_update=False
)

def plot_earthquakes(min_mag):

    filtered_df = df_geo[
        df_geo["Magnitude"] >= min_mag
    ]

    filtered_df = filtered_df.copy()

    filtered_df["Marker_Size"] = (
        filtered_df["Magnitude"] - filtered_df["Magnitude"].min()
    ) * 10 + 8

    print(
        f"Displayed earthquakes: {len(filtered_df)}"
    )

    fig = px.scatter_map(
        filtered_df,
        lat="Latitude",
        lon="Longitude",
        color="Depth",
        size="Marker_Size",
        size_max=30,
        hover_data=[
            "Magnitude",
            "Depth",
            "Date(UTC)"
        ],
        color_continuous_scale="Viridis_r",
        zoom=3,
        center={
            "lat": -35,
            "lon": -71
        },
        title=f"Earthquakes in Chile (Magnitude ≥ {min_mag:.1f})"
    )

    fig.update_traces(
        marker=dict(
            sizemin=4
        )
    )

    fig.update_layout(
        template="plotly_white",
        height=700,
        coloraxis_colorbar_title="Depth (km)"
    )

    fig.show()


# Output area
output = widgets.interactive_output(
    plot_earthquakes,
    {"min_mag": min_magnitude}
)

# Show map first, controls below
display(
    widgets.VBox([
        output,
        min_magnitude
    ])
)
```

**Analysis**

Earthquakes are distributed along the entire Chilean territory, following the Pacific coast and the Andes region. Larger earthquakes appear concentrated in specific seismic zones, while depth varies considerably across locations. Deeper earthquakes are generally observed inland, whereas shallower events are more common near the coast and offshore areas, reflecting the subduction process between the Nazca and South American tectonic plates.
